# Feature Engineering

Feature engineering creates new variables that help improve analysis and understanding.

Examples include:

• Active cases  
• Death rate  
• Recovery rate

```python
# Create engineered features

df["active_cases"] = df["confirmed"] - df["deaths"] - df["recovered"]

df["death_rate"] = (df["deaths"] / df["confirmed"]) * 100

df["recovery_rate"] = (df["recovered"] / df["confirmed"]) * 100


df.head()
```

# Exploratory Data Analysis (EDA)

EDA helps uncover patterns and trends in COVID data.

The analysis includes:

• Distribution of confirmed cases  
• Distribution of deaths  
• Distribution of recovered cases  
• Trend analysis

## Distribution of Confirmed Cases

```python
plt.hist(df["confirmed"], bins=20)

plt.title("Distribution of Confirmed COVID Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Frequency")

plt.show()
```

## Distribution of Death Cases

```python
plt.hist(df["deaths"], bins=20)

plt.title("Distribution of COVID Deaths")
plt.xlabel("Deaths")
plt.ylabel("Frequency")

plt.show()
```

## Distribution of Recovered Cases

```python
plt.hist(df["recovered"], bins=20)

plt.title("Distribution of Recovered Cases")
plt.xlabel("Recovered")
plt.ylabel("Frequency")

plt.show()
```

# Trend Analysis

```python
plt.plot(df["confirmed"])

plt.title("Confirmed Cases Trend")
plt.xlabel("Time")
plt.ylabel("Confirmed Cases")

plt.show()
```

```python
plt.plot(df["recovered"])

plt.title("Recovered Cases Trend")
plt.xlabel("Time")
plt.ylabel("Recovered Cases")

plt.show()
```

# Relationship Analysis

## Confirmed vs Deaths

```python
plt.scatter(df["confirmed"], df["deaths"])

plt.title("Confirmed Cases vs Deaths")
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")

plt.show()
```

## Confirmed vs Recovered

```python
plt.scatter(df["confirmed"], df["recovered"])

plt.title("Confirmed Cases vs Recovered")
plt.xlabel("Confirmed Cases")
plt.ylabel("Recovered")

plt.show()
```

#  Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title("COVID Dataset Correlation Matrix")

plt.show()
```
